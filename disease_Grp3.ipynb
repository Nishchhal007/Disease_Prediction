What model could we use?


This is a supervised learning problem and a classification problem too.

# Logistic Regression

```python
# from sklearn.model_selection import train_test_split
# X=df.drop('prognosis',axis=1)
# Y=df['prognosis']
# x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=101)
# y_test

x_train=df.drop('prognosis',axis=1)
y_train=df['prognosis']
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
pred=log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
log_acc=accuracy_score(y_test,pred)
log_acc
```

```python
coeff=log.coef_
coeff
```

These coeff are weights of the features , we can apply feature reduction on them to reduce over fitting.

# Decission Tree



```python
from sklearn.tree import DecisionTreeClassifier
dis=DecisionTreeClassifier(random_state=42)
dis.fit(x_train,y_train)
pred=dis.predict(x_test)
```

```python
dis_acc=accuracy_score(y_test,pred)
dis_acc
```

**Random foreset**

```python
from sklearn.ensemble import RandomForestClassifier
ran_for=RandomForestClassifier(n_estimators=100)
ran_for.fit(x_train,y_train)
pred=ran_for.predict(x_test)
ran_acc=accuracy_score(y_test,pred)
ran_acc
```

```python
#SVM
```

```python
from sklearn.svm import SVC
svm=SVC(kernel="rbf")
svm.fit(x_train,y_train)
pred=svm.predict(x_test)
svm_acc=accuracy_score(y_test,pred)
svm_acc
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(10, 6))
models = ['log', 'dis', 'ran_for', 'svm']
test_accuracy = [log_acc,dis_acc,ran_acc,svm_acc]
plt.plot(models, test_accuracy, '-o')
plt.ylim(0.90, 1.05)
plt.ylabel("Accuracy Score")
plt.show()
```

we below code we are droping some features depending of thier weights 

```python
coeff = np.mean(log.coef_,axis=0)
threshold = np.quantile(np.abs(coeff), q=0.25) 
import plotly.express as px
from plotly.offline import plot, iplot, init_notebook_mode
init_notebook_mode(connected=True)

fig = px.bar(
    x=coeff,
    y=x_train.columns,
    orientation='h',
    color=coeff,
    color_continuous_scale=[(0, 'red'), (1, 'blue')],
    labels={'x': "Coefficient Value", 'y': "Feature"},
    title="Feature Importance From Model Weights"
)

fig.update_yaxes(showticklabels=False)
fig.add_vline(x=threshold, line_color='green')
fig.add_vline(x=-threshold, line_color='green')
fig.add_vrect(x0=threshold, x1=-threshold, line_width=0, fillcolor='green', opacity=0.2)

fig.show()
```

```python
low_importance_features = x_train.columns[np.abs(coeff) < threshold]
low_importance_features
```

```python
df_new=df.drop(low_importance_features,axis=1)
```

```python
x_test.drop(low_importance_features,axis=1,inplace=True)
```
